```python
#drops all columns that contain null values
df.dropna(axis=1)
```

```python
df.notnull().sum()
```

```python
#Drops all rows hase less than number of non null values
df.dropna(axis=1,thresh=92)
```

```python
#Replace all null values by x
df.fillna("x")
```

```python
#COnvert the datatype of series to float
df['Fare'].astype(float)
```

```python
#Replace all values equal to 1 with 'one' and 0 with 'zero'
df["Parch"].replace([0,1],['zero','one'])
```
